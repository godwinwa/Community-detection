```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
#create and visualize the graph
G = nx.Graph()
G.add_edges_from([[1,2], [2,3], [1,3], [1,4], [1,5], [3,5],
                  [6,7], [7,8], [8,9], [6,9], [6,8], [7,9],
                  [10,11], [11,12], [10,12],
                  [9,12], [2,7]])
nx.draw(G, with_labels=True, node_color='r')
```

```python
#calculate and sort edge betweenness
edge_betweenness = nx.edge_betweenness_centrality(G).items()
sorted(edge_betweenness, key=lambda pair: -pair[1])[0][0]
```

```python
#remove the highest edge betweeness node to seperate communities
NUM_ITERATIONS = 3
for i in range(NUM_ITERATIONS):
    edge_betweenness = nx.edge_betweenness_centrality(G).items()
    edge_tot_delete = sorted(edge_betweenness, key=lambda pair: -pair[1])[0][0]
    G.remove_edge(*edge_tot_delete)
    nx.draw(G, with_labels=True, node_color='r')
    plt.title('Step %s\nEdge %s Deleted'%(i, edge_tot_delete))
    plt.show()
    
```
